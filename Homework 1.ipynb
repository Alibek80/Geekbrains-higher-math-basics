Напишите код на Python, реализующий построение графиков:
- окружности,
- эллипса,
- гиперболы.

```python
def my_circle(x0, y0, r):
    x = np.linspace(x0 - r, x0 + r, 1000)
    y1 = np.sqrt(r**2 - (x - x0)**2) + y0
    y2 = - np.sqrt(r**2 - (x - x0)**2) + y0
    plt.plot(x, y1, color = 'blue')
    plt.plot(x, y2, color = 'blue')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.draw()
```

```python
my_circle(5,7,3)
```

```python
def my_ellipse(x0, y0, a, b):
    x = np.linspace(x0 - a, x0 + a, 1000)
    y1 = b * np.sqrt((1 - (x - x0)**2)/(a**2)) + y0
    y2 = - b * np.sqrt((1 - (x - x0)**2)/(a**2)) + y0
    plt.plot(x, y1, color = 'green')
    plt.plot(x, y2, color = 'green')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.draw()
```

```python
my_ellipse(5,7,3,2)
```

```python
def my_hyperbola(x0, y0, a, b):
    y = np.linspace(y0 - b,y0 + b,201)
    x1 = a * np.sqrt((1+(y-y0)**2/(b**2)))+x0
    x2 = -a * np.sqrt((1+(y-y0)**2/(b**2)))+x0
    plt.plot(x1, y, color = 'green')
    plt.plot(x2, y, color = 'green')
    # plt.gca().set_aspect('equal', adjustable='box')
    plt.draw()
```

```python
my_hyperbola(5,7,3,2)
```

- Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1

- Решите систему уравнений и неравенств:

y = x2 – 1

exp(x) + x∙(1 – y) > 1

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 +1, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (0,0))

print(x1,y1)
```
